# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
# Your code here. Use apply and lambda

def revised_price(price, engine):
  for p in engine:
    if engine=='front':
      new_price = price
    else:
      new_price = 2*int(price)
    return new_price
```

```python

df['new_price'] = df.apply(lambda x: revised_price(x['price'], x['engine-location']),axis=1)
```

# <b> Problem 10 : Create a new column which stores the number of doors in a car as integers. 

```python
# Your code here. Use apply
print(df['num-of-doors'].value_counts())
def door_int(numofdoors):
  for i in [numofdoors]:
    if i=='four' :
      numofdoors = 4
    else:
      numofdoors = 2
    return numofdoors
    # print(no_of_doors)

```

```python
df['doors-int'] = df.apply(lambda x: door_int(x['num-of-doors']),axis=1)
# df['doors-int'] = df.apply(lambda x: door_int(x[),axis=1)
```
